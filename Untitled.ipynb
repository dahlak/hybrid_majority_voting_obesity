```python
def format_annotation(value):
    if value < 0:
        return f"âˆ’{abs(value):.2f}"  # Use Unicode minus sign for negative values
    elif value > 0:
        return f"+{value:.2f}"  # Use plus sign for positive values
    else:
        return f"{value:.2f}"
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection  import train_test_split
```

```python
df.head()
```

```python
X=df.drop("NObeyesdad",axis=1)
y=df["NObeyesdad"]
```

```python
#Splitting into training and testing data
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
#Standardising the Attributes
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
#####################
##### DecisionTreeClassifier
#####################
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
dec_classifier = DecisionTreeClassifier()
dec_classifier.fit(X_train,y_train)
text_visual=tree.export_text(dec_classifier)
fig=plt.figure(figsize=(25,20))
tree.plot_tree(dec_classifier,filled=True)

fig.savefig("C:\\Users\\Dah-lak\\Desktop\\Obesity\\DT.png")
```

```python
y_pred_DT=dec_classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report,accuracy_score
cm_dt = confusion_matrix(y_test, y_pred_DT)
print(cm_dt)
acc_score_DT=accuracy_score(y_test,y_pred_DT)
acc_score_DT=round(acc_score_DT*100,2)
print("Score = ",acc_score_DT)
print(classification_report(y_test, y_pred_DT))
```

```python
#####################
##### RandomForestClassifier
#####################
model=RandomForestClassifier(n_estimators=500,max_depth=5)
model.fit(X_train,y_train)
y_pred_RF=model.predict(X_test)
cm_rf = confusion_matrix(y_test, y_pred_RF)
print(cm_rf)
acc_score_RFC=accuracy_score(y_test,y_pred_RF)
acc_score_RFC=round(acc_score_RFC*100,2)
print("Score = ",acc_score_RFC)
print(classification_report(y_test, y_pred_RF))
```

```python
#####################
##### Naive Bayes Classifier
#####################
from sklearn.naive_bayes import GaussianNB
nb= GaussianNB()
nb.fit(X_train, y_train)
pred_nb = nb.predict(X_test)
cm_nb = confusion_matrix(y_test, pred_nb)
print(cm_nb)
acc_score_nb=accuracy_score(y_test,pred_nb)
acc_score_nb=round(acc_score_nb*100,2)
print("SCore = ", acc_score_nb)
print(classification_report(y_test, pred_nb))
```

```python
#####################
##### MLPClassifier
#####################
from sklearn.neural_network import MLPClassifier
nn_classifier = MLPClassifier(hidden_layer_sizes=(150,100,50), max_iter=5000,activation = 'relu',solver='adam',random_state=0)
nn_classifier.fit(X_train, y_train)
pred_nn = nn_classifier.predict(X_test)
cm_nn = confusion_matrix(y_test, pred_nn)
print(cm_nn)
acc_score_nn=accuracy_score(y_test,pred_nn)
acc_score_nn=round(acc_score_nn*100,2)
print("score: ",acc_score_nn)
print(classification_report(y_test,pred_nn))
```

```python
#####################
##### SVM
#####################
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
svc_classifier = SVC(kernel = 'rbf')
svc_classifier.fit(X_train, y_train)

pred_svc = svc_classifier.predict(X_test)
cm_svc = confusion_matrix(y_test, pred_svc)
print(cm_svc)
acc_score_svc=accuracy_score(y_test,pred_svc)
acc_score_svc=round(acc_score_svc*100,2)
print("score: ",acc_score_svc)
print(classification_report(y_test,pred_svc))
```

```python
#####################
##### LogisticRegression
#####################
from sklearn.linear_model import LogisticRegression
log_classifier = LogisticRegression()
log_classifier.fit(X_train, y_train)

pred_log = log_classifier.predict(X_test)

cm_log = confusion_matrix(y_test, pred_log)
print(cm_log)
acc_score_log=accuracy_score(y_test,pred_log)
acc_score_log=round(acc_score_log*100,2)
print("score: ",acc_score_log)
print(classification_report(y_test,pred_log))
```

```python
#####################
##### XGBClassifier
#####################
from xgboost import XGBClassifier
xg_classifier = XGBClassifier()
xg_classifier.fit(X_train, y_train)
pred_xg = xg_classifier.predict(X_test)

cm_xg = confusion_matrix(y_test, pred_xg)
print(cm_xg)
acc_score_xg=accuracy_score(y_test,pred_xg)
acc_score_xg=round(acc_score_xg*100,2)
print("score: ",acc_score_xg)
print(classification_report(y_test,pred_xg))
```

```python
#####################
##### GradientBoostingClassifier
#####################
from sklearn.ensemble import GradientBoostingClassifier
grad_classifier = GradientBoostingClassifier()
grad_classifier.fit(X_train, y_train)
pred_grad = grad_classifier.predict(X_test)

cm_grad = confusion_matrix(y_test, pred_grad)
print(cm_grad)
acc_score_grad=accuracy_score(y_test,pred_grad)
acc_score_grad=round(acc_score_grad*100,2)
print("score: ",acc_score_grad)
print(classification_report(y_test,pred_grad))
```

```python
#####################
##### KNeighborsClassifier
#####################
from sklearn.neighbors import KNeighborsClassifier
knn_classifier = KNeighborsClassifier(n_neighbors = 5)
knn_classifier.fit(X_train, y_train)
pred_knn = knn_classifier.predict(X_test)

cm_knn = confusion_matrix(y_test, pred_knn)
print(cm_knn)
acc_score_knn=accuracy_score(y_test,pred_knn)
acc_score_knn=round(acc_score_knn*100,2)
print("score: ",acc_score_knn)
print(classification_report(y_test,pred_knn))
```
